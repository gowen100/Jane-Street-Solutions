```python
import numpy as np
import time
import requests
from bs4 import BeautifulSoup
import functools
```

```python
url='https://www.janestreet.com/puzzles/what-about-bob/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
y =[text for text in soup.body.stripped_strings]
print("Puzzle")
print("~~~~~~")
print(" ".join(y[7:11]))
```

```python
@functools.lru_cache(maxsize=None)
def bob_wins(target,last,choices):
    poss = [i for i in choices if i<=target and i != last]
    if target in poss:
        return False

    else:
        if any([bob_wins(target-p,p,choices) == True for p in poss]):
            return False
        else:
            return True
      
start = time.time()    
win_count = 0
n=0
while win_count < 10:
    n += 1
    if bob_wins(n,0,tuple([i+1 for i in range(9)])):
        win_count += 1
        print("For N = {} Bob wins ({} mod 32)".format(n,n % 32))
print("Solved in {:,.4f} seconds".format(time.time()-start))
```
