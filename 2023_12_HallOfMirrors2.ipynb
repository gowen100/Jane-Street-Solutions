```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display,HTML,Image
```

```python
url='https://www.janestreet.com/puzzles/hall-of-mirrors-2-index/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
display(Markdown('### Puzzle as published \n'+url))
y =[text for text in soup.body.stripped_strings]
#display([(i,j) for i,j in enumerate(y)])
display(Markdown("### "+y[13]))
display(Image('https://www.janestreet.com/puzzles/20231201_mirrors_2.PNG',width=400))
```
